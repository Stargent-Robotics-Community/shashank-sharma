```python
import pyttsx3
import speech_recognition as sr
import requests
import json
import datetime
import os
import wikipedia


# trying asking hello , how are you , what you can do for me , do you have a girlfriend , what do you like the most , do you like web series . This bot is able to play music , opening spotify , opening microsoft edge , wikipedia .
engine = pyttsx3.init()
voices = engine.getProperty('voices')
engine.setProperty('rate' , 180)
voice = engine.setProperty('voices' , voices[0])

def speak(audio) :
    print(audio)
    engine.say(audio)
    engine.runAndWait()


def take_command():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("listening...")
        audio = r.listen(source)

    try :
        query = r.recognize_google(audio , language='en-in')
        print(query)
        return query.lower()

    except Exception as e :
        print(e)


def wish_me():
    hours = int(datetime.datetime.now().hour)
    if hours >= 0 and hours < 12 :
        speak("Good morning Buddy...!")
        
        
    elif hours >= 12 and hours < 18 :
        speak("Good evening buddy...!")
        
    else:
        speak("Good night .. my friend .. ")

def task_master(query):
    print("bot :" , end="")
    if " play music" in query :
        speak("ok .... presenting the list of songs which song do you want me to play...")
        path_m = "E:\\songs"
        music = os.listdir(path_m)
        for index , item in enumerate(music):
            print(index , item)
        a = int(take_command())
        os.startfile(os.path.join(path_m , music[a]))
    
    elif "open spotify"  in query:
        print("bot :" , end="")
        speak("ok ... opening spotify")
        path_wt = 'C:\\Users\\user\\Desktop\\Spotify.lnk'
        os.startfile(path_wt)
    
    elif "open edge" in query:
        print("bot :" , end="")
        speak("opening microsoft edge...")
        path_ed = 'C:\\Program Files (x86)\\Microsoft\\Edge\\Application\\msedge.exe'
        os.startfile(path_ed)

    elif "wikipedia" in query:
        print("bot :" , end="")
        speak(" ok  ... what do you want to search on wikipedia")
        que = take_command().lower()
        results = wikipedia.summary(que , sentences=3)
        speak("after searching on wikipedia i have found this....")
        print(results)
        speak(results)

    elif "quit" in query :
        quit()
if __name__ == "__main__":
    wish_me()
    direction = {'hello':"hey there ..." , 'how are you':"fine how are you ??" , "i am fine too" :"nice to hear that" , "i am fine":"good to hear that" , "i am good" :"nice to hear that" , "what you can do for me":"I can do many things for you" , "bye": "byee for today may we met again" , "do you have a girlfriend": "nope", "do you like memes": "yes i do like memes" , "what do you like the most":" i like many things like dancing , singing. what do you like..?" , "do you watch web series":"yes i do watch web series" , "do you watch movies":"yeah for sure"  }
    while(True):
        task = take_command()
        if task in direction.keys():
            print("bot :" , end="")
            speak(direction[task])
        else :
            task_master(task)
```
